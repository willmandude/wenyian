# 问题三

```python
import matplotlib.pyplot as plt
import talib as ta
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data =  pd.read_excel('sz50.xlsx', sheetname = '600036.XSHG',index_col='datetime')
data_dict = data['close']
```

```python
a = (ta.MA(data_dict,10)[:])
```

```python
plt.figure(figsize=(15, 7))
plt.plot(data_dict)#线图
plt.plot(a)#线图
plt.ylabel('Price')
plt.show()
```
